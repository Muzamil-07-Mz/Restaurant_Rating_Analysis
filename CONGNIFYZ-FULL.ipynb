```python
price_range_ratings = dataset.groupby('Price range')['Aggregate rating'].mean()

print("AVERAGE RATINGS FOR EACH PRICE RANGE:\n")
print(price_range_ratings)
print("\n")

highest_avg_rating_color = price_range_ratings.idxmax()
print("THE AVERAGE RATING IS:",highest_avg_rating_color)
```

```python
price_range_ratings = {}

for price_range in dataset['Price range'].unique():
    price_range_dataset = dataset[dataset['Price range'] == price_range]
    avg_rating = price_range_dataset['Aggregate rating'].mean()
    price_range_ratings[price_range] = avg_rating

max_avg_rating = max(price_range_ratings.values())
best_price_range = [price_range for price_range, avg_rating in price_range_ratings.items() if avg_rating == max_avg_rating][0]

best_color = dataset.loc[dataset['Price range'] == best_price_range, 'Rating color'].mode().values[0]

print("THE COLOR THAT REPRESENTS THE HIGHEST AVERAGE RATING AMONG DIFFERENT PRICE RANGES IS:", best_color)
```

```python
plt.bar(price_range_ratings.keys(), price_range_ratings.values())
max_idx = list(price_range_ratings.values()).index(max_avg_rating)
plt.bar(list(price_range_ratings.keys())[max_idx], max_avg_rating, color='orange')
plt.show()
```

# LEVEL 3:

# TASK 1: Predictive Modeling

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x = dataset[['Average Cost for two',
                   'Has Table Booking',
                   'Has Online Delivery',
                   'Is delivering now',
                   'Switch to order menu',
                   'Price range',]] 
x
```

```python
y=dataset['Aggregate rating']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg_model=LinearRegression()
lin_reg_model.fit(x_test,y_test)

lin_reg_pred=lin_reg_model.predict(x_test)

lin_reg_accuracy=int(round(lin_reg_model.score(x_test,y_test)*100))

print("MODEL ACCURACY OF LINEAR REGRESSION IS : ",lin_reg_accuracy,"%")

mse = mean_squared_error(y_test, lin_reg_pred)
print("MEAN SQUARED ERROR OF MODEL LINEAR REGRESSION IS :", mse)

r2 = r2_score(y_test, lin_reg_pred)
print("R SQUARED SCORE OF MODEL LINEAR REGRESSION IS :", r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
dcs_tr_model=DecisionTreeRegressor(random_state=10)
dcs_tr_model.fit(x_train, y_train)

dcs_tr_pred=dcs_tr_model.predict(x_test)

dcs_tr_accuracy=int(round(dcs_tr_model.score(x_test,y_test)*100))

print("MODEL ACCURACY OF DECISION TREE IS : ",dcs_tr_accuracy,"%")

mse = mean_squared_error(y_test, dcs_tr_pred)
print("MEAN SQUARED ERROR OF MODEL DECISION TREE IS :", mse)

r2 = r2_score(y_test, dcs_tr_pred)
print("R SQUARED SCORE OF MODEL DECISION TREE IS :", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rnd_fr_model=RandomForestRegressor(n_estimators=100, random_state=10)
rnd_fr_model.fit(x_train, y_train)

rnd_fr_pred=rnd_fr_model.predict(x_test)

rnd_fr_accuracy=int(round(rnd_fr_model.score(x_test,y_test)*100))

print("MODEL ACCURACY OF RANDOM FOREST IS : ",rnd_fr_accuracy,"%")

mse = mean_squared_error(y_test, rnd_fr_pred)
print("MEAN SQUARED ERROR OF MODEL RANDOM FOREST IS :", mse)

r2 = r2_score(y_test, rnd_fr_pred)
print("R SQUARED SCORE OF MODEL RANDOM FOREST IS :", r2)
```
